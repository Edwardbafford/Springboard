In the above code we have verified our ability to resize and reshape images into rows. We will next iterate through all the images we have a create a large 2D array which will be saved to a CSV.

We now clean all of the data and save it into CSV files.

```python
import os
import random

#Helper functions used below
def start(result_array, size, num):
    result_array = np.empty((0, size))
    num = num + 1
    return result_array, num
    
def end(result_array, num):
    np.savetxt(path + 'Cleaned\\images' + str(random.randint(1,101)) + str(num) + '.csv', result_array, delimiter=",")
    print('Saved ' + str(num))
```

```python
num = 1000 #Num needs to be 1000 so randomization of the files works properly
size = 13245
rescale = .08
csv_size = 50
result_array = None

result_array, num = start(result_array, size, num) 

for key in sub.keys():
    for index, file in enumerate(os.listdir(os.fsencode(path + sub[key]))):
        #Clean image
        img = io.imread(path + sub[key] + '\\' + str(os.fsdecode(file)), as_grey=True)
        small = transform.rescale(img, rescale)
        row = small.reshape(-1)
        result = np.append(row, key)
        result_array = np.append(result_array, [result], axis=0)
        
        print(os.fsdecode(file))
        
        #Periodically Save
        if ((index + 1) % csv_size) == 0:
            end(result_array, num)
            result_array, num = start(result_array, size, num)
    
    #Save after switching directories
    end(result_array, num)
    result_array, num = start(result_array, size, num)
```

Finally we verify that the saved images are properly configured by randomly checking a couple

```python
#Helper function for testing
def test_csv(file, h, l):
    df = pd.read_csv(path + 'Cleaned\\' + file, header=None)
    viewer.ImageViewer(df.iloc[random.randint(0,49),:-1].values.reshape(h, l)).show()
```
