## Games
Collect all regular season team matchups and outcomes from 2009 to 2017

```python
import requests
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
#Get games played
r = requests.post('https://profootballapi.com/schedule?api_key=9SLf6UbjIWVYevtA5Tmi1HuphNK7Gr08')
games = pd.DataFrame(r.json())
games.head()
```

```python
#Filter out pre-season games, teams aren't always trying to win so it will introduce poor data
games2 = games[games['season_type'] == 'REG']
games2.season_type.value_counts()
```

```python
#Filter out current 2018 season. Information is incomplete.
games3 = games2[games2.year != '2018']
games3.year.value_counts()
```

```python
#Get diffence and binary win columns
games3['diff'] = games3['home_score'] - games3['away_score']
games3['home_win'] = (games3['diff'] > 0).astype(int)
games3.head()
```

```python
#Filter out unnecessary rows
games_final = games3[['home','home_score','away','away_score','diff','home_win','id','year','week']]
games_final.head()
```
