## Team Stats
Collect and clean team level statistics for each team each year

```python
import requests
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
#Get team stats per game
r = requests.post('https://profootballapi.com/teams?api_key=9SLf6UbjIWVYevtA5Tmi1HuphNK7Gr08&season_type=REG')
teams = pd.DataFrame(r.json())
teams.head()
```

```python
#Get games df to merge and obtain yearly information
games = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\games.csv')[['id','year']]
games.head()
```

```python
#Merge for year and filter important columns
teams2 = games.merge(teams, left_on='id', right_on='nfl_game_id')[['id','year','team','pyds','ryds','totfd','trnovr','pen']]
teams2.head()
```

```python
#Collect yearly statistics through groupby
teams_final = teams2.groupby(['year','team'])['pyds','ryds','totfd','trnovr','pen'].sum().reset_index()
teams_final.head()
```

```python
#Save to csv
teams_final.to_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\teams.csv',index=False)
```
