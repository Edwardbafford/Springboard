```python
import requests
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
#Get rushing players
rushing = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\rushing.csv')
r_players = rushing.groupby(['year','team','name']).sum().reset_index()
r_players.head()
```

```python
#Get recieving players
recieving = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\recieving.csv')
c_players = recieving.groupby(['year','team','name']).sum().reset_index()
c_players.head()
```

```python
#Get passing players
passing = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\passing.csv')
p_players = passing.groupby(['year','team','name']).sum().reset_index()
p_players.head()
```

```python
#Get player stats to write to
player_stats = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\players.csv')
player_stats.head()
```

```python
#Get QB run to pass ratio
top_qb = p_players.merge(p_players.groupby(['year','team'])['yards'].max().reset_index(),on=['year','team','yards'])
qb_full = top_qb.merge(r_players,on=['year','team','name'])

player_stats['qb_pass_to_run'] = qb_full['attempts_x'].divide(qb_full['attempts_x'] + qb_full['attempts_y']).round(2)
player_stats.head()
```

```python
#Get RB run to recieve ratio
top_rb = r_players.merge(r_players.groupby(['year','team'])['yards'].max().reset_index(),on=['year','team','yards'])

rb_full = top_rb.merge(c_players,on=['year','team','name'],how='left')
rb_full.fillna(0)

player_stats['rb_run_to_catch'] = rb_full['attempts'].divide(rb_full['attempts'] + rb_full['receptions']).round(2)
player_stats.head()
```

```python
#Save to csv
player_stats.to_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\players.csv',index=False)
```

```python
#Get team stats
team_stats = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\teams.csv')
team_stats.head()
```

```python
#Merge all the stats
all_stats = team_stats.merge(player_stats,on=['team','year'])
all_stats.head()
```

```python
#Get games
games = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\games.csv')
games.head()
```

```python
home_all = games.merge(all_stats,left_on=['year','home'],right_on=['year','team'])
wrangle_all = home_all.merge(all_stats,left_on=['year','away'],right_on=['year','team'],suffixes=('_home','_away'))
wrangle_all.head()
```
