## Rushing Stats
Aggregate individual rushing stats by game from the passing csv into player stats per year for each team. Meant to be run directly after pashing notebook.

```python
import requests
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
#Get rushing df
rushing = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\rushing.csv')
rushing.head()
```

```python
#Get player stats to write to
player_stats = pd.read_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\players.csv')
player_stats.head()
```

```python
#Get rushing stats for the team
rush_agg = rushing.groupby(['year','team'])['attempts','touchdowns'].sum().reset_index()
rush_agg.head()
```

```python
#Merge with other player stats
player_stats2 = player_stats.merge(rush_agg, on=['year','team'])
player_stats2.head()
```

```python
#Transform for important columns
player_stats2['pass_to_run'] = player_stats2['pass_attempts'].divide(player_stats2['pass_attempts'] + player_stats2['attempts']).round(2)
player_stats3 = player_stats2.rename(columns={'touchdowns': 'rush_touchdowns'})
player_stats4 = player_stats3.iloc[:,[0,1,9,3,4,5,6,8]]
player_stats4.head()
```

```python
#Get average of top rusher per team
rush_players = rushing.groupby(['year','team','name'])['attempts','yards'].sum().reset_index()
top_rusher = rush_players.groupby(['year','team'])['yards'].max().to_frame().reset_index()
top_rusher2 = rush_players.merge(top_rusher, on=['year','team','yards'])
top_rusher2['avg'] = top_rusher2['yards'].divide(top_rusher2['attempts']).round(2)
player_stats4['running_back_avg'] = top_rusher2['avg']
player_stats4.head()
```

```python
#Save to csv
player_stats4.to_csv('C:\\Users\\Louie\\Repositories\\Springboard\\Capstone 1 - NFL Matchup Predictions\\Data Wrangling\\Data\\players.csv',index=False)
```
